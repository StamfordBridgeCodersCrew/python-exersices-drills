```python
technologies   = ({
    'Courses':["Spark","PySpark","Hadoop","Python","Pandas","Hadoop","Spark","Python","NA"],
    'Fee' :[22000,25000,23000,24000,26000,25000,25000,22000,1500],
    'Duration':['30days','50days','55days','40days','60days','35days','30days','50days','40days'],
    'Discount':[1000,2300,1000,1200,2500,None,1400,1600,0]
})

df = pd.DataFrame(technologies)
df
```

```python

'''
1. Dataframe :- Group by multiple columns
2. Apply aggregation

'''
grouped_df = df.groupby(by=["Courses", "Duration"]).sum()
grouped_df
```

```python
'''
1. Dataframe :- Group by multiple columns
2. Apply aggregation

'''

'''
You can also choose whether to include NA/None/Nan in group keys or not
by setting dropna parameter.

To not to include None/Nan values on group keys 
Set dropna=False

By default , tis set true
'''
grouped_df = df.groupby(by=["Courses", "Duration"], dropna=False).sum()
grouped_df
```

```python

'''
1. Dataframe :- Groupby column
2. Apply multiple aggregations on all (numeric dtype columns) of the grouped_df

'''
aggregation_list = ["min", "max", "count"]
answer_df = df.groupby(by="Courses").aggregate(aggregation_list)
answer_df
```

```python
'''
1. Dataframe :- Groupby column
2. Apply multiple aggregations on some (numeric dtype columns) of the grouped_df
'''
aggregation_dict = {
    "Duration": "count",
    "Fee": ["min", "max"],
    "Discount": ["min", "max", "mean"]
}

answer_df = df.groupby(by="Courses").aggregate(aggregation_dict)
answer_df
```

```python
'''
1. Dataframe :- Groupby column
2. Apply multiple aggregations on any 1 (numeric dtype columns) of the grouped_df
'''

aggregation_list = ["min", "max", "mean"]
answer_df = df.groupby(by="Courses")["Fee"].aggregate(aggregation_list)
answer_df
```

## **[Replace NaN values with 0 in single, multiple columns of the df](https://sparkbyexamples.com/pandas/pandas-replace-nan-values-by-zero-in-a-column/)**

```python
import pandas as pd
import numpy as np

technologies = {
    'Courses': ["Spark",np.nan,"Hadoop","Python","pandas",np.nan,"Java"],
    'Fee' : [20000,25000, np.nan,22000,24000,np.nan,22000],
    'Duration': [np.nan,'40days','35days', np.nan,'60days','50days','55days'],
    'Discount': [1000,np.nan,1500,np.nan,2500,2100,np.nan]
}

df = pd.DataFrame(technologies)
df
```

```python

'''
1. Replace NaN Values with 0
   On a single column

df["col"] = df["col"].fillna(value=0.0)
df["col"] = df["col"].replace(to_replace=np.nan, value=0.0)

2. Replace NaN Values with 0
   On multiple columns

df[ ["col1", "col2", "col3"] ] = df[ ["col1", "col2", "col3"] ].fillna(value=0.0)
df[ ["col1", "col2", "col3"] ] = df[ ["col1", "col2", "col3"] ].replace(to_replace=np.nan, value=0.0)

3. Replace NaN Values with 0
   On all columns

new_df = df.fillna(value=0.0)
new_df = df.replace(to_replace=np.nan, value=0.0, inplace=False)

4. Replace NaN Values with 0
   On the entire pandas dataframe 

new_df = df.replace(to_replace=np.nan, value=0.0, inplace=False)
'''

df["Fee"] = df["Fee"].fillna(value=0.0)
#df["Fee"] = df["Fee"].replace(to_replace=np.nan, value=0.0)
df
```

```python

df[ ["Fee","Duration"] ] = df[ ["Fee","Duration"] ].fillna(value=0.0)
#df[ ["Fee","Duration"] ] = df[ ["Fee","Duration"] ].replace(to_replace=np.nan, value=0.0)
df
```

```python

'''
1. Replace NaN Values with 0
   On a single column

df["col"] = df["col"].fillna(value=0.0)
df["col"] = df["col"].replace(to_replace=np.nan, value=0.0)

2. Replace NaN Values with 0
   On multiple columns

df[ ["col1", "col2", "col3"] ] = df[ ["col1", "col2", "col3"] ].fillna(value=0.0)
df[ ["col1", "col2", "col3"] ] = df[ ["col1", "col2", "col3"] ].replace(to_replace=np.nan, value=0.0)

3. Replace NaN Values with 0
   On all columns

new_df = df.fillna(value=0.0)
new_df = df.replace(to_replace=np.nan, value=0.0, inplace=False)

4. Replace NaN Values with 0
   On the entire pandas dataframe 

new_df = df.replace(to_replace=np.nan, value=0.0, inplace=False)
'''

new_df = df.replace(to_replace=np.nan, value=0.0, inplace=False)
new_df
```

# **[Sorting column values in Pandas](https://sparkbyexamples.com/pandas/pandas-dataframe-sort_values/)**

```python
import pandas as pd
import numpy as np
technologies = ({
    'Courses':["Spark",np.nan,"pandas","Java","Spark"],
    'Fee' :[20000, 18000, 18000, 22000, 22000],
    'Duration':['30days','40days','35days','60days','50days'],
    'Discount':[1000, 2500, 1500, 1200, 3000]
})

#df = pd.DataFrame(technologies, index = ['r1','r3','r5','r2','r4'])
df = pd.DataFrame(data=technologies)
df
```

```python
'''
SELECT *
FROM df
ORDER BY Courses ASC
'''

'''
If axis = 0 :- Then sort by columns 
If axis = 1 :- Then sort by rows/indices
'''
# Sort DataFrame by column Fee in ASC order
new_df1 = df.sort_values(by="Fee", axis=0)

# Sort DataFrame by column "Fee" in DESC order
new_df2 = df.sort_values(by="Fee", axis=0, ascending=False)

new_df1
```

```python

'''
By default, NaN  values are pushed at the end of the DataFrame

To push NaN values to the start of the DataFrame :- na_position="first"
'''
new_df3 = df.sort_values(by="Courses", na_position="first")
```

```python

# Sort DataFrame by column Fee in ASC order , Then
# Sort DataFrame by column Discount in DESC order , Then
new_df4 = df.sort_values(by=["Fee", "Discount"], ascending=[True, False], na_position="last")
new_df4

# Sort DataFrame by column Fee in DESC order , Then
# Sort DataFrame by column Discount in ASC order , Then
new_df5 = df.sort_values(by=["Fee", "Discount"], ascending=[False, True], na_position="last")
new_df5

# Sort DataFrame by column Fee in DESC order , Then
# Sort DataFrame by column Discount in DESC order , Then
new_df6 = df.sort_values(by=["Fee", "Discount"], ascending=[False, False], na_position="last")
new_df6
```
